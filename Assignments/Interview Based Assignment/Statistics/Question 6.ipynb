**Using Shapiro-Wilk test**

To determine if the change in blood pressure follows a normal distribution in the given clinical trial, we can use a hypothesis test called the **Shapiro-Wilk test**. 

This test assesses the null hypothesis that the data is normally distributed.

```python
from scipy.stats import shapiro
```

```python
# Extract the 'Blood Pressure Before' and 'Blood Pressure After' columns
blood_pressure_before = df['Blood Pressure Before']
blood_pressure_after = df['Blood Pressure After']

# Calculate the change in blood pressure
change_in_blood_pressure = blood_pressure_after - blood_pressure_before

# Perform Shapiro-Wilk test
statistic, p_value = shapiro(change_in_blood_pressure)

# Print the results
print(f"Shapiro-Wilk Test Statistic {statistic:.2f}")
print(f"p-value: {p_value:.2f}")

# Interpret the results
alpha = 0.05  # significance level

if p_value > alpha:
    print("The change in blood pressure follows a normal distribution.")
else:
    print("The change in blood pressure does not follow a normal distribution.")
```

**Interpretation:**

The Shapiro-Wilk test assesses the null hypothesis that the data follows a normal distribution.

- If the p-value is greater than the chosen significance level (typically 0.05), we fail to reject the null hypothesis, suggesting that the data is normally distributed.
- Conversely, if the p-value is less than or equal to the significance level, we reject the null hypothesis, indicating that the data does not follow a normal distribution.

**Therefore, since p-value is less than 0.05 we reject the null hypothesis which means data does not follow normal distribution.**

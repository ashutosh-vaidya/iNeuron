```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read data from the csv file downloaded from the above link
df = pd.read_csv("Electric_Vehicle_Population_Data.csv")
df.head()
```

**Get all the cars and their types that do not qualify for clean alternative fuel vehicle**

```python
# create a column clean_fuel
df1 = df.copy()
df1["clean_fuel"] = df["Clean Alternative Fuel Vehicle (CAFV) Eligibility"] != "Clean Alternative Fuel Vehicle Eligible"
df1.loc[df1["clean_fuel"] == True]
```

**Get all TESLA cars with the model year, and model type made in Bothell City.**

```python
df1 = df[df['City'] == 'Bothell']
df2 = df1[df1['Make'] == 'TESLA']
df2
```

**Get all the cars that have an electric range of more than 100 and were made after 2015**

```python
electric_cars = df[(df['Electric Range'] > 100) & (df['Model Year'] > 2015)][['Make', 'Electric Vehicle Type']]
print("Cars with electric range > 100 made after 2015:")
print(electric_cars)
print()
```

**Draw plots to show the distribution between city and electric vehicle type**

```python
plt.figure(figsize=(12, 6))
df.groupby(['City', 'Electric Vehicle Type']).size().unstack().plot(kind='bar', stacked=True)
plt.xlabel('City')
plt.ylabel('Count')
plt.title('Distribution between City and Electric Vehicle Type')
plt.legend(title='Electric Vehicle Type')
plt.xticks(rotation=45)
plt.show()
```
